# Problem 1: Mobile Recharge Validation System
 Problem Statement

Create a function to validate a mobile recharge request.

Rules:
Recharge amount must be ≥ ₹50


Valid plans: 199, 299, 399, 599


User should be allowed to retry if invalid


Real-Time Use

Telecom recharge apps


Payment gateways


Hint

Use function for validation


Use while loop for retry


```python
#function to validate recharge
def validate_recharge(amount, plans):
  if amount <50:
    print("Recharge amount should be greater than or equal to ₹50")
    return False
  elif amount in plans:
    return True
  else:
    print("Invalid recharge plan")
    return False
```

```python
valid_plans=[199,299,399,599]
print("valid plans are:",valid_plans)
while True:
  amount=int(input("Enter the recharge amount: "))
  if validate_recharge(amount,valid_plans):
    print(f"Recharge of {amount} is successful")
    break
```

# Problem 2: Inventory Reorder Alert System
Problem Statement


Given a dictionary of products and stock quantities:

Print Reorder Alert if stock < 15


Print Stock OK otherwise


Real-Time Use

Warehouse systems


Retail management


Hint

Use function


Use for loop on dictionary


```python
#Reorder Alert system function
def alert_system(product,stock):
    if stock<15:
      print(f"Reorder Alert: {product} stock is {stock}")
    else:
      print(f"{product} stock is OK")
```

```python
products={
    "laptop":10,
    "Mobiles":17,
    "Speakers":15,
    "Sound Bars":12,
    "Printers":6
}
for product,stock in products.items():
  alert_system(product,stock)
```

# Problem 3: Student Result Processing System
Problem Statement

Create a function that:

Accepts a list of student marks


Calculates average marks


Prints Pass if average ≥ 50 else Fail


Real-Time Use
Online exam systems


Evaluation portals


Hint

Use loop to calculate total


Return result from function





```python
#Result processing Function
def result_processing(marks):
  result=sum(marks)/len(marks)
  if result >=50:
    return result,"Pass"
  else:
    return result,"Fail"
```

```python
marks=[45,89,76,47,58,56]
result,status=result_processing(marks)
print("Student Marks:",marks)
print("Average Marks:",result)
print("Status:",status)
```

# Problem 4: Cab Fare Estimator with Retry Option
Problem Statement

Create a cab fare calculator:

Base fare = ₹50


₹12 per km


Peak hour → 25% extra


Allow retry using while loop


Real-Time Use

Ride-hailing apps


Transport systems


Hint

Function for fare calculation


While loop for retry



```python
#fare calculation for a ride
def fare_estimator(distance,peak_hour):
  base_fare=50
  fare_per_km=12
  fare=(distance*fare_per_km)
  if fare<=base_fare:
    fare=base_fare
  if peak_hour:
    fare*=1.25
  return fare
```

```python
while True:
  distance=float(input("Enter the distance in km: "))
  peak_hour=input("Is it a peak hour? (yes/no): ").lower()
  if peak_hour=="yes":
    peak_hour=True
  else:
    peak_hour=False

  fare=fare_estimator(distance,peak_hour)
  print("Estimated Fare:",fare)

  choice=input("Do you want to calculate another fare? (yes/no)").lower()
  if choice!="yes":
    break
```

# Problem 5: Employee Attendance Eligibility Checker
Problem Statement

Create a function that:

Accepts attendance list ("P", "A")


Calculates attendance percentage


Returns Eligible if ≥ 75% else Not Eligible


Real-Time Use

HR payroll systems


Internship tracking


 Hint

Use loop to count present days


Use return statement




```python
#Attendance Eligibility checking function
def attendance_checker(attendance):
  present_days=0
  total_days=len(attendance)
  for i in attendance:
    if i=="P":
      present_days+=1
  attendance_percentage=(present_days/total_days)*100
  if attendance_percentage>=75:
    return "Eligible"
  else:
    return "Not Eligible"
```

```python
attendance=["P","P","P","A","P","P","P","P","P","P"]
result=attendance_checker(attendance)
print(result)
```

# Problem 6: Password Strength Checker
Problem Statement

Create a function to check password strength.

Rules:

Minimum length = 8


Must contain at least one digit


Must contain at least one special character (@#$)


Real-Time Use

Login systems


Security validation


Hint

Use loop over string


Use conditions inside function


```python
from string import digits
#Function To validate password
def password_strength_checker(password):
  min_len=8
  has_digit=False
  has_special_char=False
  if len(password)<min_len:
    print("Password must be at least 8 characters long")
    return False
  for char in password:
    if char in "@#$":
      has_special_char=True
    if char in digits:
      has_digit=True
  if has_special_char and has_digit:
    print("Password Matches Requirements")
    return True
  elif has_digit==True and has_special_char==False:
    print("Password does not contain Special Character'@','#','$'")
    return False
  elif has_digit==False and has_special_char==True:
    print("Password does not contain digit")
    return False
  elif has_digit==False and has_special_char==False:
    print("Password does not contain digits and special characters")
  else:
      print("Password meets All requirements")
      return True
```

```python
password=input("Enter the password: ")
while True:
  if password_strength_checker(password):
    break
  else:
    password=input("Enter the new password: ")
```
